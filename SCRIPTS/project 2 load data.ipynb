```python
import pandas as pd
import yfinance as yf

# Download raw daily data
tickers = ["CL=F", "JETS", "SPY"]

raw = yf.download(
    tickers,
    start="2010-01-01",
    end="2025-01-01",
    auto_adjust=True,
    progress=False
)

print("Raw downloaded data preview:")
print(raw.head())
print("\nRaw shape:", raw.shape)
print("\nRaw columns:")
print(raw.columns)

#Save raw data exactly as downloaded
raw.to_csv("raw_yfinance_daily_data.csv")

print("\nSaved raw file:")
print("raw_yfinance_daily_data.csv")
```
